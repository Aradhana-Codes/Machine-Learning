```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("Position_Salaries.csv")
```

```python
data
```

```python
x=data.iloc[:,1:2].values
y=data.iloc[:,2].values
x
y
```

```python
#Polynomial Regression with degree 2
poly_regs=PolynomialFeatures(degree=2)
x_poly=poly_regs.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lin_reg_2.predict(poly_regs.fit_transform(x)),color="red")
plt.title("Polynomial Regression (Degree 2)")
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

```python
#predicting result
poly_pred=lin_reg_2.predict(poly_regs.fit_transform([[6.5]]))
```

```python
print("Predicted Salary for level 6.5 (degree 2):",poly_pred)
```

```python
#polynomial regression with degree 3
poly_regs=PolynomialFeatures(degree=3)
x_poly=poly_regs.fit_transform(x)
lin_reg_3=LinearRegression()
lin_reg_3.fit(x_poly,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lin_reg_3.predict(poly_regs.fit_transform(x)),color="red")
plt.title("Polynomial Regression (Degree 3)")
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

```python
#predicting result
poly_pred=lin_reg_3.predict(poly_regs.fit_transform([[6.5]]))
```

```python
print("Predicted Salary for level 6.5 (degree=3)",poly_pred)
```
